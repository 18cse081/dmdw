```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
path=('https://raw.githubusercontent.com/18cse005/DMDW/main/USA_cars_datasets.csv')
```

```
data = pd.read_csv(path)
```

```
import seaborn as sns
```

```
data.head()
```

```
data.dropna(axis=0,inplace=True)# removed the null values 1st method rows when large data we are having
data.isnull().sum()
```

data visualization


```
plt.figure(figsize=(20,8))
plt.scatter(data['price'],data['mileage'],c='green')
plt.xlabel('price')
plt.ylabel('mileage')
plt.show()
```

According to the graph as per the price is concerned according to this the mileage of the cars are varis

```
plt.figure(figsize=(25,10))
plt.hist(data['brand'],edgecolor='white',bins=25)
plt.title('Histogram of Mileage ')
plt.xlabel('mileage')
plt.ylabel('price')
plt.show()
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['price'],y=data['mileage'])
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['price'],y=data['mileage'],marker='*',fit_reg=False)
```

```
sns.distplot(data['price'])
```

```
sns.displot(data['price'],kde=False)
```

```
sns.displot(data['price'],kde=False,bins=10)
```

```
plt.figure(figsize=(25,10))
sns.countplot(x=data['year'],data=data)
```

```
plt.figure(figsize=(25,10))
sns.countplot(x=data['country'],data=data,hue='state')
```

```
sns.boxplot(y='price',data=data)
```

```
plt.figure(figsize=(15,6))
sns.distplot(data.price[data.mileage==0])
sns.distplot(data.price[data.mileage==1])
plt.legend(['0','1'])
plt.show()
```
